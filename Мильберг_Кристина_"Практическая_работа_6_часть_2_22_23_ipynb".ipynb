### Задание 7

На вертикальной столбиковой диаграмме для самого популярного (popularity) и самого непопулярного жанров покажите средние значения по всем характеристикам. Отрицательные и положительные значения характеристик показывайте разными цветами.

```r
library(dplyr)
library(tidyr)
library(ggplot2)

music <- music %>% 
  select(-duration_ms)
most_popular_genre <- music %>% 
  group_by(music_genre) %>% 
  summarize(mean_popularity = mean(popularity, na.rm = TRUE)) %>% 
  slice_max(mean_popularity) %>% 
  pull(music_genre)
means_df <- music %>%
  filter(music_genre == most_popular_genre) %>%
  select(acousticness:valence) %>%
  summarise_all(~mean(., na.rm = TRUE)) %>%
  pivot_longer(cols = everything(), names_to = "Характеристики", values_to = "Средние_значения")

if(length(most_popular_genre) == 0){
  message("No data found for the most popular genre.")
}else{
  means_df$palette <- ifelse(means_df$Средние_значения > 0, "blue", "red")

ggplot(means_df, aes(x = Характеристики, y = Средние_значения, fill = palette)) +
  geom_col() +
  scale_fill_manual(values = c("blue", "red")) +
  theme_classic() +
  labs(title = paste("Средние значения характеристик для жанра", most_popular_genre))}
```
